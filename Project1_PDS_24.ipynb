```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv')
df
```

```python
null_columns = df.columns[df.isnull().any()]
print("Columns with null values:", null_columns.tolist())
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Cabin'].fillna('Unknown', inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='Sex', data=df, palette=['blue', 'red'])
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.title("Survival Count by Gender")
plt.legend(title="Gender")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.title("Survival Count by Passenger Class")
plt.legend(title="Passenger Class", labels = ['1st Class', '2nd Class', '3rd Class'])
plt.show()
```

```python
def categorize_age(age):
    if age <= 16:
        return 0
    elif age <= 26:
        return 1
    elif age <= 36:
        return 2
    elif age <= 62:
        return 3
    else:
        return 4

df['AgeClass'] = df['Age'].apply(categorize_age)
df
```

```python
df.drop(columns=['Age'], inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='AgeClass', data=df)
plt.xticks([0, 1], ['Dead', 'Survived'])
plt.xlabel("Survival Status")
plt.ylabel("Count")
plt.title("Survival Count by Age Class")
plt.legend(title="Passenger Class")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df['Number of Relatives'] = df['SibSp'] + df['Parch']

alive_passengers = df[df['Survived'] == 1]

plt.figure(figsize=(10, 6))
plt.scatter(alive_passengers['Number of Relatives'], alive_passengers['Age'])
plt.xlabel('Number of Relatives')
plt.ylabel('Age')
plt.title('Scatter Plot of Survived Passengers Age vs. Number of Relatives')
plt.grid(True)
plt.show()
```
